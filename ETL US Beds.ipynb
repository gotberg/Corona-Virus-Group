```python
# Import/ Extract a datafile, dataset
# pd.read_csv("name of file")
us_beds_df = pd.read_csv("hospital_beds_USA_v1.csv")
us_beds_df
```

```python
# look up column datatypes
us_beds_df.dtypes
```

Transform us_beds DataFrame

```python
# Transform
# Drop/ Eliminate unnecessary columns
# out of 12 original columns drop all, except for state, lat, lng, type, beds, population, year
us_beds_df= us_beds_df.drop(columns=['country', 'county', 'measure', 'source', 'source_url'], axis = 1)
us_beds_df
```

```python
# querry for all dataframes (all columns) but filtered by 'year'
us_beds_2019= us_beds_df.loc[us_beds_df['year']== 2019]
us_beds_2019
```

```python
# Jamie's recommendation first state then year
# state_us_beds_df[['state','beds','population']].groupby(['state']).sum()
```

```python
# only keep population count from 2019
# us_beds_count_df= us_beds_df[['state','beds','population']].groupby(['state']).sum()
# us_beds_count_df
```

```python
# sum beds by state for 2019
# us_beds_count_df= us_beds_df[['state','beds','population']].groupby(['state']).sum()
state_beds_2019= us_beds_2019[['state', 'beds', 'population']].groupby(['state']).sum()
state_beds_2019
```

```python
# reset index for sum of beds by state for 2019
indexed_us_beds_count_df= state_beds_2019.reset_index()
indexed_us_beds_count_df
```
